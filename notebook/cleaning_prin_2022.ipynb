Proviamo a filtrare i progetti nel campo delle DH

```python
parole_chiave_dh = [
    #digital x
    "digital humanities", "digital archive", "digital preservation", "digital heritage",
    "digital transition", "digital mapping", "digital catalogue", "digital repertoire",
    "digital atlas", "digitisation", "metadata", "digital repository",
    "digital enhancement", "archival infrastructures", "digital curation", "digital restoration",
    "digital linguistics", "digital linguistic", "digital cartography",
    "digital reconstruction",  "virtual heritage",
    "computer vision", "digital broadcasting archives", "3D modeling",
    "digital dissemination"

    # LINGUISTICA COMPUTAZIONALE E ANALISI TESTUALE
    "computational linguistics", "computational linguistic",
    "textual", "corpus", "semantic", "syntax", "lexical", "philology", 
    "scholarly edition", "encoding", "paratext", "data-driven", "data oriented", 
    "computational", "algorithmic", "machine learning", "human-machine interaction",
    "distributional semantic analysis", "linked data", "interlinking", "text mining",
    "named entity recognition", "stylometry", "digital papyrology", "scholarly digital edition",
    "digital epigraphy", "formal models", "conceptual models",

    # --- GEOGRAFIA DIGITALE, CARTOGRAFIA E ARCHEOLOGIA ---
    "GIS", "preventive archaeology", "spatial analysis", "deep mapping",
    "historical GIS",

    "knowledge graph", "digital journalism", "data journalism",
]


parole_chiave_sh = [
    # Economia e Finanza
    "economics", "gdp", "inflation", "unemployment", "taxation", "bank", "banking", "economy",

    # Sociologia, Antropologia e Demografia
    "sociology", "demography", "migration", "inequality", "gender", 
    "ethnicity", "anthropological", "ritual", "voting",

    # Scienze Politiche, Istituzioni e Relazioni Internazionali
    "politics", "political", "democracy", "diplomacy", "institution", 
    "legislation", "regulation", "treaty",

    # Psicologia e Scienze Cognitive
    "psychology", "psychological", "cognition", "perception", 
    "anxiety", "depression", "personality",

    # Geografia Umana e Media
    "urbanization", "pedagogy"
]
```

```python
indici_progetti_dh = []

for index, row in prin_2022.iterrows():
    progetto = row["progetto"].lower()
    
    is_dh = any(chiave_dh in progetto for chiave_dh in parole_chiave_dh)
    is_sh = any(chiave_sh in progetto for chiave_sh in parole_chiave_sh)

    if is_dh and not is_sh:
        indici_progetti_dh.append(index)
```

```python
df_progetti_dh = prin_2022.loc[indici_progetti_dh].reset_index(drop=True)

df_progetti_dh
```

```python
print(f"Trovati {len(df_progetti_dh)} progetti di Digital Humanities:")

for index, row in df_progetti_dh.iterrows():
    print(f"{index+1} -> {row["progetto"]}")
```

```python
finanziamenti_progetti_DH = df_progetti_dh["finanziamento"].sum()
int(finanziamenti_progetti_DH)
```
